```python
from mif import *
mif = load_mif()
print(mif)
```

```python
mtxlist = [mif.gadget,mif.karateclub,mif.erdosReny,mif.scalefree,mif.homophilly,mif.heterophilly]
[adjacencyinfocheck(i) for i in mtxlist]
adjacencylist = ['gadget', 'karateclub', 'erdosReny', 'scalefree', 'homophilly', 'heterophilly']
```

```python
for i in range(0,len(mtxlist)):
    print("")
    print(f"MiF example for {adjacencylist[i]}")
    print("")
    log = MiF(mtxlist[i], 4, 32, 0.5, 3)
    print(log)
    print("...")
    del log
print("")

for i in range(0,len(mtxlist)):
    print("")
    print(f"MiFbroadcast example for {adjacencylist[i]}")
    print("")
    log = MiF_broadcast(mtxlist[i], 3)
    print("...")
    del log
print("")

for i in range(0,len(mtxlist)):
    print("")
    print(f"MiFbroadcast example for {adjacencylist[i]}")
    print("")
    log = MiF_broadcast(mtxlist[i], 3, loop = 1)
    print("...")
    del log
print("")


for i in range(0,len(mtxlist)):
    print("")
    print(f"MiFDI for {adjacencylist[i]}")
    print("")
    log = MiFDI(mtxlist[i])
    print("...")
    del log
print("")

for i in range(0,len(mtxlist)):
    print("")
    print(f"MiFDI for {adjacencylist[i]}")
    print("")
    log = MiFDI(mtxlist[i], startingvertices="max")
    print("...")
    del log
print("")

for i in range(0,len(mtxlist)):
    print("")
    print(f"MiFDI for {adjacencylist[i]}")
    print("")
    log = MiFDI(mtxlist[i], loop = 1)
    print("...")
    del log
print("")

for i in range(0,len(mtxlist)):
    print("")
    print(f"MiFDI for {adjacencylist[i]}")
    print("")
    log = MiFDI(mtxlist[i], startingvertices="max", loop =1)
    print("...")
    del log
```
