```python
#This was done on Google Colab.
def save_sparse_to_json(matrix, path):
    if not issparse(matrix):
        matrix = csr_matrix(matrix)
    coo = matrix.tocoo()
    data = {
        "row": coo.row.tolist(),
        "col": coo.col.tolist(),
        "data": coo.data.tolist(),
        "shape": list(coo.shape)
    }
    with open(path, "w", encoding="utf-8") as f:
        json.dump(data, f)
save_sparse_to_json(adjacencyinfocheckedlist[1],"/content/drive/My Drive/Colab Notebooks/karate.json")
karatejson = fileOnColab("karate.json")
adjacencyinfocheck(karatejson)
```
